```python
import pandas as pd
import numpy as np
import seaborn as sns
import calendar
import matplotlib.pyplot as plt
import random

%autosave 1
```

```python
raw_df = pd.read_csv("Happiness 2019.csv")
raw_df
```

```python
#null value analysis
ax = plt.axes()
ax.set_facecolor("white")
raw_df.isna().sum().plot(kind='bar', color='red', figsize=(15, 7), fontsize=12)
plt.xlabel('Columns', labelpad=7, loc='center', fontsize=15)
plt.ylabel('Missing count', labelpad=7, fontsize=15)
plt.title('Missing information analysis', pad=15, fontsize=25)
plt.grid(False)
```

```python
country_or_region = raw_df.groupby('Country or region').aggregate({'Country or region':'count'})
country_or_region.rename(columns={'Country or region':'Country or region count'},inplace=True)
country_or_region.sort_values('Country or region count',ascending=False)
#here found that the data has no repeatative country 
```

```python
df = raw_df.copy(deep=True)
df.set_index('Country or region',inplace=True)
df
```

```python
raw_df[raw_df['Country or region']=='India']
#getting the data of india
```

```python
df['Positivity Score'] = df[df.columns[1:-1]].sum(axis=1)
df.sort_values('Positivity Score',ascending=False)
# so here we can conclude that the best country to live is norway
```

```python
df['Best country score'] = df['Positivity Score']-df['Perceptions of corruption']
```

```python
#shifting thenegative column to the end
column = df.pop('Perceptions of corruption')
df.insert(9,'Perceptions of corruption', column)
df
```

```python
for x in df.columns:
    df[x].sort_values().plot(kind='barh', figsize=(10, 30), color='#ffcc00')
    plt.grid(False)
    plt.title(f"2019 {x}", fontsize=20,pad=15)
    plt.xlabel(x, fontsize=15)
    plt.ylabel('Country', fontsize=15)
    plt.show();
```

```python
df[df.columns[1:-1]].reset_index().sort_values('Best country score').plot(x='Country or region', kind='barh', stacked=True, figsize=(15, 30));
plt.grid(False)
plt.title(f"2019 Happiness", fontsize=20,pad=15)
plt.xlabel('Overall data', fontsize=15)
plt.ylabel('Country', fontsize=15)
plt.show();
```
